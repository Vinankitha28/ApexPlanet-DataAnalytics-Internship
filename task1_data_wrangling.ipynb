```python
import pandas as pd
import numpy as np

# Create sample ecommerce dataset
np.random.seed(42)
n = 1000

df = pd.DataFrame({
    'Order_ID': ['ORD' + str(i).zfill(4) for i in range(1, n+1)],
    'Customer_ID': ['CUST' + str(np.random.randint(1, 200)).zfill(3) for _ in range(n)],
    'Customer_Name': np.random.choice(['Amit', 'Priya', 'Rahul', 'Sneha', 'Vijay', 'Deepa', 'Ravi', 'Anita'], n),
    'Category': np.random.choice(['Electronics', 'Clothing', 'Books', 'Home & Kitchen', 'Sports'], n),
    'Sub_Category': np.random.choice(['Mobile', 'Laptop', 'Shirt', 'Novel', 'Cookware', 'Shoes'], n),
    'Product_Name': np.random.choice(['iPhone 14', 'Dell Laptop', 'Nike Shoes', 'Python Book', 'Samsung TV', 'Mixer'], n),
    'Order_Date': pd.date_range(start='2023-01-01', periods=n, freq='8H'),
    'Ship_Date': pd.date_range(start='2023-01-03', periods=n, freq='8H'),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], n),
    'City': np.random.choice(['Mumbai', 'Delhi', 'Bangalore', 'Chennai', 'Hyderabad'], n),
    'Sales': np.random.uniform(100, 50000, n).round(2),
    'Quantity': np.random.randint(1, 10, n),
    'Discount': np.random.choice([0, 0.1, 0.2, 0.3], n),
    'Profit': np.random.uniform(-500, 10000, n).round(2),
    'Payment_Mode': np.random.choice(['Credit Card', 'UPI', 'Net Banking', 'COD'], n),
    'Customer_Age': np.random.randint(18, 65, n),
    'Rating': np.random.choice([1, 2, 3, 4, 5], n)
})

# Add some missing values to practice cleaning
df.loc[np.random.choice(df.index, 50), 'Customer_Age'] = np.nan
df.loc[np.random.choice(df.index, 30), 'Rating'] = np.nan

print("Dataset created! Shape:", df.shape)
print("Columns:", df.columns.tolist())
df.head()
```

```python
# Save to use across all tasks
df.to_csv('ecommerce_data.csv', index=False)
print("Dataset saved!")

# Download to your computer
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv('ecommerce_data.csv')
print("✅ Dataset loaded!")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
df.head()
```

```python
# Cell 3 - Data Dictionary
data_dict = {
    'Order_ID':       'Unique ID for each order',
    'Customer_ID':    'Unique ID for each customer',
    'Customer_Name':  'Name of the customer',
    'Category':       'Product category',
    'Sub_Category':   'Product sub category',
    'Product_Name':   'Name of the product',
    'Order_Date':     'Date when order was placed',
    'Ship_Date':      'Date when order was shipped',
    'Region':         'Geographic region of sale',
    'City':           'City of delivery',
    'Sales':          'Revenue from the order',
    'Quantity':       'Number of items ordered',
    'Discount':       'Discount applied on order',
    'Profit':         'Profit earned from order',
    'Payment_Mode':   'How customer paid',
    'Customer_Age':   'Age of the customer',
    'Rating':         'Customer rating 1 to 5'
}

dict_df = pd.DataFrame(list(data_dict.items()), 
                        columns=['Column Name', 'Description'])
print("✅ Data Dictionary Created!")
print(dict_df.to_string(index=False))
```

```python
# Cell 4 - Data Quality Assessment
print("=" * 50)
print("MISSING VALUES")
print("=" * 50)
print(df.isnull().sum())

print("\n" + "=" * 50)
print("DUPLICATE ROWS")
print("=" * 50)
print("Total Duplicates:", df.duplicated().sum())

print("\n" + "=" * 50)
print("DATA TYPES")
print("=" * 50)
print(df.dtypes)

print("\n" + "=" * 50)
print("BASIC STATISTICS")
print("=" * 50)
df.describe()
```

```python
# Cell 5 - Fix Missing Values
print("Before cleaning:")
print("Missing values:", df.isnull().sum().sum())

# Fix numerical columns with median
df['Customer_Age'].fillna(df['Customer_Age'].median(), inplace=True)
df['Rating'].fillna(df['Rating'].median(), inplace=True)

print("\nAfter cleaning:")
print("Missing values:", df.isnull().sum().sum())
print("✅ Missing values fixed!")
```

```python
# Cell 6 - Remove Duplicates
before = len(df)
df.drop_duplicates(inplace=True)
after = len(df)

print(f"Rows before: {before}")
print(f"Rows after:  {after}")
print(f"Removed:     {before - after} duplicate rows")
print("✅ Duplicates removed!")
```

```python
# Cell 7 - Fix Date Formats
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Ship_Date']  = pd.to_datetime(df['Ship_Date'])

print("Order_Date type:", df['Order_Date'].dtype)
print("Ship_Date type: ", df['Ship_Date'].dtype)
print("✅ Dates fixed!")
```

```python
# Cell 8 - Feature Engineering (create new useful columns)
df['Order_Year']      = df['Order_Date'].dt.year
df['Order_Month']     = df['Order_Date'].dt.month
df['Order_Month_Name']= df['Order_Date'].dt.strftime('%B')
df['Delivery_Days']   = (df['Ship_Date'] - df['Order_Date']).dt.days
df['Profit_Margin']   = (df['Profit'] / df['Sales'] * 100).round(2)
df['Age_Group']       = pd.cut(df['Customer_Age'], 
                                bins=[0,25,35,50,100], 
                                labels=['Youth','Young Adult','Adult','Senior'])

print("✅ New columns created!")
print("New columns:", ['Order_Year','Order_Month','Order_Month_Name',
                       'Delivery_Days','Profit_Margin','Age_Group'])
df.head()
```

```python
# Cell 9 - Handle Outliers in Sales
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['Sales'] < lower) | (df['Sales'] > upper)]
print(f"Outliers found: {len(outliers)}")

df = df[(df['Sales'] >= lower) & (df['Sales'] <= upper)]
print(f"Clean dataset shape: {df.shape}")
print("✅ Outliers handled!")
```

```python
# Cell 10 - Save Cleaned Dataset
df.to_csv('cleaned_ecommerce_data.csv', index=False)
print("✅ Cleaned dataset saved as cleaned_ecommerce_data.csv")
print("Final shape:", df.shape)
```
